```python
import random
import numpy as np


first_dice = []
element_50 = []

for i in range(0,50):
    element_50.append(i)
    first_dice.insert(i, random.randint(1,6))

second_dice = []
for i in range(0,50):
    second_dice.insert(i, random.randint(1,6))

sum_dice_50 = [first_dice[i] + second_dice[i] for i in range(len(second_dice))]
```

```python
import seaborn as sns
plot = sns.lineplot(x =element_50, y=sum_dice_50, marker="o",lw=1)
plot.set_xlabel("Order of casting two dices")
plot.set_ylabel("Sum")
```

```python
sns.histplot(x=sum_dice_50, kde=True, discrete=True)
```
